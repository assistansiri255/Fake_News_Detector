```python
import numpy as np
import pandas as pd
import sklearn
import joblib
import itertools
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import PassiveAggressiveClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
print(np.__version__)
print(pd.__version__)
print(sklearn.__version__)
```

```python
df=pd.read_csv(r"C:\Users\dp580\Downloads\news.csv.zip")  
print(df.shape)    #shape tell number of rows and columns
df.head()
```

```python
labels=df.label
labels.head()
```

```python
# split the data into training and testing data
x_train,x_test,y_train,y_test=train_test_split(df['text'],labels,test_size=0.2,random_state=7)
```

```python
# from sklearn.feature_extraction.text import TfidfVectorizer
# from sklearn.linear_model import PassiveAggressiveClassifier
# from sklearn.metrics import accuracy_score

# initialize tfidfvectorizer
tfidf_vectorizer = TfidfVectorizer(stop_words="english", max_df=0.7)

# fit and transform on train data
tfidf_train = tfidf_vectorizer.fit_transform(x_train)

# **transform on test data**
tfidf_test = tfidf_vectorizer.transform(x_test)

# train classifier
pac = PassiveAggressiveClassifier(max_iter=50)
pac.fit(tfidf_train, y_train)

# predict on test
y_pred = pac.predict(tfidf_test)

# accuracy
score = accuracy_score(y_test, y_pred)
print(f'Accuracy: {round(score*100, 2)}%')
```

```python
print(confusion_matrix(y_test,y_pred,labels=['FAKE','REAL']))
print(classification_report(y_test, y_pred))
```

```python
# Save model and vectorizer
joblib.dump(pac, "fake_news_model.pkl")
joblib.dump(tfidf_vectorizer, "tfidf_vectorizer.pkl")
```

```python
import joblib

# Load saved model and vectorizer
pac = joblib.load("fake_news_model.pkl")
tfidf_vectorizer = joblib.load("tfidf_vectorizer.pkl")

# Now you can predict new news
news = ["Breaking news: something unbelievable happened!"]
news_vector = tfidf_vectorizer.transform(news)
prediction = pac.predict(news_vector)
print(prediction[0])  # Output: FAKE or REAL
```
